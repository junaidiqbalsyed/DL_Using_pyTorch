# Gradient Descent 

```python
# Def function 

def fx(x) :
    return np.cos( 2* np.pi * x) + x**2

#Â Derivative of the function 

def deriv(x) :
    return ( -1 * (np.sin(2 * np.pi * x) * 2 * np.pi ) )+ 2*x
```

```python
#Â define the range of x 


x = np.linspace( -2 , 2 , 2001 )
plt.figure(figsize= ( 10,10 )) 
plt.plot(x , fx(x) )
plt.plot(x , deriv(x) )
plt.xlim(x[[0 , -1]])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
# Random starting point 

localmin = np.random.choice(x , 1) 

print(f" The initial local minima is {localmin} ")

# Learning parameters 
learning_rate = 0.01 
epochs = 100 

# Trainig cycle 
for i in range(epochs): 
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad
    
print(f" the final local minima is {localmin} ")
```

```python
# Plot the results 


plt.figure(figsize= ( 10,10 )) 
plt.plot(x , fx(x) )
plt.plot(x , deriv(x) )
plt.plot(localmin , fx(localmin) , "*" )
plt.plot(localmin , deriv(localmin) , "*" )
plt.xlim(x[[0 , -1]])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
#Â Lets try to plot the gradients at every epoch 


# Random starting point 

localmin = np.random.choice(x , 1) 
print(f" Initial local minima is {localmin} ")

# Learning parameters 
learning_rate = 0.01 
epochs = 100

# array that has the learning parameters for both the function and the derivatives 
params = np.zeros((epochs,2)) 

# Trainig cycle 
for i in range(epochs): 
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad
    params[i, :] = localmin, grad
    
print(f" the final local minima is {localmin} ")

# Plot the dradients for every iterations 

plt.figure(figsize= (18,7 ))
plt.plot(params[:,0] , "r*", label =  "local minima" )
plt.plot(params[:,1] , "b.-", label = "Derivatives" )
plt.legend()
plt.grid()
plt.show()
```

### The gola of the gradient descent is to move towards the derivative of 0 

From the large derivative it moves towards zero. when the derivative is zero thats when we are at the local minima 

```python
# Ploting the derivaives on the graph 

# Plot the results 


plt.figure(figsize= ( 15,10 )) 
plt.plot(x , fx(x) )
plt.plot(x , deriv(x) )

plt.plot(params[:,0] ,  [ fx(i) for i in params[:,0]] , "r*" )
plt.plot(params[:,1] ,  [ deriv(i) for i in params[:,1]] , "g*" )

plt.xlim(x[[0 , -1]])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

# Artificial Neural Networks 

#### Building a neural network using pytorch 

```python
import numpy as np 
import torch 
import torch.nn as nn
import matplotlib.pyplot as plt 
```

```python
# Create the data 

N = 300
x = torch.randn(N,1)
y = x + torch.randn(N,1)/2

# plot the values 
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
# Build the model 

model = nn.Sequential(
    nn.Linear(1,1),   # input layer
    nn.ReLU(),        # activation function
    nn.Linear(1,1)    # output layer
)

model
```

```python
# learning rate 
lr = 0.05 

# loss function 
lossfun = nn.MSELoss()

# optimizer 
optimizer = torch.optim.SGD(model.parameters() , lr = lr)
```

```python
# train the model 

epochs = 400
losses = torch.zeros(epochs)

# Train the model !

for e in range(epochs):
    
    # forwarp prop
    yHat = model(x)
    
    # compute loss
    loss = lossfun(yHat , y)
    losses[e] = loss
    
    # backprop
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# plot the errors

plt.plot(losses.detach().numpy())
plt.show()
```

#### Building a neural network using Keras

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
k_model = Sequential()
k_model.add(Dense( 20 , activation= 'tanh', input_shape = (1, ) ))
k_model.add(Dense( 12 , activation= 'tanh', input_shape = (1, ) ))
k_model.add(Dense( 1 ) )

k_model.summary()
```

```python
k_model.compile(loss = 'mean_squared_error', optimizer='adam',  metrics=["mean_squared_error"])
```

```python
his = k_model.fit(x.detach().numpy(),y.detach().numpy() ,  batch_size=32, epochs= 50, validation_split=0.1)
```

```python
k_model.evaluate( x.detach().numpy(),y.detach().numpy()  )
```
